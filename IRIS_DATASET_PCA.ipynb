```python
# Creation  of : Features Matrix and Target Array
```

```python
X_iris = iris.drop('species', axis=1)
X_iris.shape
```

```python
y_iris = iris['species']
y_iris.shape
```

# PCA :  Iris Dimensionality

```python
# Step1: choose model class

from sklearn.decomposition import PCA
```

```python
# Step2: Instantiate the model with hyperparameters

model = PCA(n_components = 2)
```

```python
# Step3: Fit to data. ***Note: y is not specified***

model.fit(X_iris)
```

```python
# Step4: Transform the data to two dimensions

X_2D = model.transform(X_iris)
```

```python
# Visualization
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]
sns.lmplot("PCA1", "PCA2", hue='species', data=iris, fit_reg=False);
```

# Iris Clustering

```python
# Step1: choose model class

from sklearn.mixture import GaussianMixture
```

```python
# Step2: Instantiate the model with hyperparameters

model = GaussianMixture(n_components=3,
           covariance_type = 'full')
```

```python
# Step3: Fit to data. ***Note: y is not specified***

model.fit(X_iris)
```

```python
# Step4: Determine cluster labels

y_gmm = model.predict(X_iris)
```

```python
# Visualization
iris['cluster'] = y_gmm
sns.lmplot('PCA1','PCA2', data = iris, hue = 'species',
          col = 'cluster', fit_reg = False);
```
